```python
def slicer (time,Load_resist):

    
#Try finding the closest peaks of the force signal based on the peaks of the resistances.
#makes sure there is an association between the peaks of the forces and peaks of resistance.

#no peak of res -- no force --> false peak

    flag = False
    start_inx = [] 
    end_inx =[] 

    j = 0

    for i in range(len(Load_resist)):
        if (Load_resist[i] <0.004)& (flag == False):
            continue
        elif (Load_resist[i] >= 0.004)&(flag== False): 
            if (j >= 1 ) : 
                if ( (i- end_inx[j - 1] ) > 150):
                    start_inx.append(i)
                    flag = True
                    j = j + 1
            else:
                start_inx.append(i)
                flag = True
                j = j + 1

        elif (Load_resist[i] >= 0.004)&( flag == True):
            continue
        elif (Load_resist[i] <0.004)&( flag == True) & ((i- start_inx[j -1]) > 240 ):
            end_inx.append(i)
            flag = False

#     if len(start_inx) > 10:
#         del end_inx[-1]
#         del end_inx[-1]
#         end_inx.append(start_inx[len(start_inx)-1])
#         del start_inx[-1]
        

    for i in range(len(start_inx)):
        piecetime = []
        pieceres = []
        peaks_amplitude = {}
        peaks_time = {}
        piecetime = time[start_inx[i]:end_inx[i]]
        pieceres = Load_resist[start_inx[i]:end_inx[i]]
#         print(Load_resist)
#         print(pieceres)
        #finding peaks 
        peaks_ind, _ = find_peaks(pieceres)
#         print(i)
#         print(peaks_ind)
        peaks_amplitude[str(i)]=pieceres.iloc[peaks_ind]
        peaks_time[str(i)] = piecetime.iloc[peaks_ind]
#         print(peaks_time[str(i)])
        plt.figure(figsize=(50, 20))
        plt.plot(piecetime,pieceres)
        plt.scatter(peaks_time[str(i)],peaks_amplitude[str(i)])

#     return (piecetime[i],pieceres[i])
```
