```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

path = "C:\\Users\\fatem\\Desktop\\University\\FYP\\FK_dummytummy\\Extra_points"
destination = "C:\\Users\\fatem\\Desktop\\University\\FYP\\FK_dummytummy\\V4"
files = os.listdir(path)
print(files)


def windows_(d, w, t):
    r = np.arange(len(d))
    s = r[::t]
    z = list(zip(s, s + w))
    f = '{0[0]}:{0[1]}'.format
    g = lambda t: d.iloc[t[0]:t[1]]
    return pd.concat(map(g, z), keys=map(f, z))


def array_convert(c, window):
    f_list = []
    for i in range(0, len(c), window):
        if i != len(c) - 1:
            x = c[i:i + window]
            if (len(x) == window):
                f_list.append(x)
    return f_list


def dump_results(path, destination):
    for i in range(0, 4):
        print("i is :",i)
        new_path = path + '\\' + files[i]
        print(new_path)
        data0 = pd.read_csv(new_path)
        print(data0)
        print(data0.shape)
        # search keyword with ''
        r1 = data0.iloc[:, 10]
        r2 = data0.iloc[:, 11]
        r3 = data0.iloc[:, 12]
        r4 = data0.iloc[:, 13]

        detrended_r = []
        r_ = [r1, r2, r3, r4]

        for d in range(0, 4):
            data = r_[d]
            window_size = 10
            stride = 1

            # print(final_list)
            windows = windows_(data, window_size, stride)
            windows = array_convert(windows, window_size)
            windows = np.array(windows)
            moving_vars = windows.var(axis=1)

            # remove nans
            # final_list = moving_vars[window_size - 1:]
            final_list = moving_vars
            # plt.plot(final_list)
            # plt.show()
            detrended_r.append(final_list)

        print(len(detrended_r[0]))
        print(len(detrended_r[1]))
        print(len(detrended_r[2]))
        print(len(detrended_r[3]))

        data_ = {'s1_d': detrended_r[0], 's2_d': detrended_r[1], 's3_d': detrended_r[2], 's4_d': detrended_r[3]}

        df = pd.DataFrame(data_)
        print(df.shape)
        df.to_csv(destination + '\\' + files[i])


if __name__ == '__main__':
    dump_results(path, destination)
```
